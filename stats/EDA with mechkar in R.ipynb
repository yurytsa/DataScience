```r
df <- read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00571/hcvdat0.csv")
```

```r
head(df)
```

# EDA

```r
exploreData(data.frame(df))
```

```r
table(df$Category)
```

```r
df$Category <- ifelse(df$Category=='0=Blood Donor' | df$Category=='0s=suspect Blood Donor',0,1)
```

```r
df$Category <- factor(df$Category, levels=c(0,1), labels=c("Normal","Disease"))
```

```r
table(df$Category)
```

```r
exploreData(data.frame(df), y="Category")
```
