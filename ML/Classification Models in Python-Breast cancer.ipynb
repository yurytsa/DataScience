# Brest Cancer Prediction

```python
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
df = pd.read_csv("../Data/Brest_cancer_dataR2.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df2 = df.copy()
df2 = df2.rename(columns={'MCP.1':'MCP1'})
df2['Classification'] = df2['Classification'].astype('category')
```

```python
sns.pairplot(df2,hue="Classification")
```

# Partition of the data

```python
from sklearn.model_selection import train_test_split
```

```python
nv = ['Age','BMI','Glucose','Insulin','HOMA','Leptin','Adiponectin','Resistin','MCP1']
yv = 'Classification'
```

```python
X = df2.loc[:,nv]
y = df2.Classification
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.40,random_state=42)
## with a seed of 2 we got unbalaced partition, we changed it to 4
```

# Partitioning using pyMechkar

```python
from pyMechkar.analysis import train_test
```

```python
train, test = train_test(data=df2, prop=0.7, seed=42)
```

```python
X_train = train.loc[:,nv]
y_train = train.Classification
X_test = test.loc[:,nv]
y_test = test.Classification
```

```python
X_train.head()
```

# Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

```python
mod1 = LogisticRegression()
mod1.fit(X_train,y_train)
```

```python
yhat1_tr = mod1.predict(X_train)
yhat1_tr
```

```python
yprob1_tr = mod1.predict_proba(X_train)
yprob1_tr
```

```python
pd.crosstab(y_train,yhat1_tr)
```

```python
### Accuracy:
mod1.score(X_train,y_train)
```

```python
### Train
mod1.score(X_train,y_train)
yprob1_tr = mod1.predict_proba(X_train)
print(roc_auc_score(y_train, yprob1_tr[:,1]))
```

```python
yhat1_ts = mod1.predict(X_test)
print(mod1.score(X_test,y_test))
print(pd.crosstab(y_test,yhat1_ts))
yprob1_ts = mod1.predict_proba(X_test)
print(roc_auc_score(y_test, yprob1_ts[:,1]))
```

```python
res = {1:{'model':'Logistic Regression',
           'accuracy-Train':mod1.score(X_train,y_train),
           'accuracy-Test':mod1.score(X_test,y_test),
           'AUC-train':roc_auc_score(y_train, yprob1_tr[:,1]),
           'AUC-test':roc_auc_score(y_test, yprob1_ts[:,1])
      }}
```

```python
res
```

```python
yhat2_ts = mod2.predict(X_test)
print(pd.crosstab(y_test,yhat2_ts))
yprob2_ts = mod2.predict_proba(X_test)
print('---------------')
print(mod2.score(X_test,y_test))
print(roc_auc_score(y_test, yprob2_ts[:,1]))
```
